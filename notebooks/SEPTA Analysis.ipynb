```python
# We want the latest readings..
url="https://www3.septa.org/hackathon/TrainView/"
d=pd.read_json(url)

# Make changes on cols
cols = ['sum','mean','max','min','count']
f = {'late':cols}
g=d[['line','late']].groupby(['line']).agg(f)
g.columns = cols
g=g.reset_index()
# g.to_json()
g.values.tolist()

```

```python
org = "30th%20Street%20Station"
des = "Elkins%20Park"
url="https://www3.septa.org/hackathon/NextToArrive/{}/{}/10".format(org,des)
# 30th to Elkins
d=pd.read_json(url)
d['orig_train']=d['orig_train'].apply(lambda x: str(x))


url="https://www3.septa.org/hackathon/Arrivals/{}/28/".format(org)
records = requests.get(url).json()
# o=pd.read_json(url)
# o.head()
records
time = list(records.keys())[0]
n = records[time][0]['Northbound']
dn=pd.DataFrame(n)

p=d.merge(dn[['train_id','track','track_change','status']], 'left', left_on=['orig_train'],
                                  right_on=['train_id'])

p[['track','orig_departure_time','status','orig_arrival_time','orig_line','orig_train','train_id']]
```
