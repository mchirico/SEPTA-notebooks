```python
import pandas as pd
```

```python
url="https://www3.septa.org/hackathon/TrainView/"
d=pd.read_json(url)
```

```python
d.head()
```

```python
cols = ['sum','mean','max','min','count']
f = {'late':cols}
g=d[['line','late']].groupby(['line']).agg(f)
g.columns = cols
g=g.reset_index()
# g.to_json()
g
```
